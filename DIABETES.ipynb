METHODS FOR HANDLING MISSING VALUES

The df.dropna(inplace=True) function removes all rows containing missing (NaN) values from the DataFrame.

1. dropna() identifies and removes rows where at least one column has a missing value.
2. inplace=True modifies the original DataFrame instead of returning a new one.

This code fills missing (NaN) values in all numeric columns with their respective median values to handle missing data efficiently.

1. df.select_dtypes(include=['number']).columns selects all numeric columns in the DataFrame.
2. df[numeric_columns].fillna(df[numeric_columns].median()) replaces NaN values with the median of each numeric column.

```python
numeric_columns = df.select_dtypes(include=['number']).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())
```

**Step#4 OUTLIER DETECTION**

This code detects outliers in numeric columns using the Z-score method:

1. Computes Z-scores to measure how far each value deviates from the mean, normalized by the standard deviation.

2. Identifies outliers as values where the absolute Z-score is greater than 3 (more than 3 standard deviations from the mean).

3. Counts the number of outliers in each numeric column and prints the results.

```python
#Outlier detection using NumPy
z_scores = (df[numeric_columns] - np.mean(df[numeric_columns].values, axis=0)) / np.std(df[numeric_columns].values, axis=0)
#Identify outliers
outliers = np.abs(z_scores) > 3
#Count outliers per column
outlier_counts = np.sum(outliers, axis=0)
print("Number of outliers per numeric column:")
for col, count in zip(numeric_columns, outlier_counts):
    print(f"{col}: {count}")
```

**Step#8  FEATURE IMPORTANCE**

Calculates the correlation between each feature and diabetes progression (Y).
*   abs() ensures we get the strength of the relationship, regardless of whether it is positive or negative.

*  sort_values(ascending=False) orders the features from most to least correlated with Y.

```python
#Compute correlation of all features with Y
correlation = df.corr()['Y'].abs().sort_values(ascending=False)

#Display correlation values
print("Feature Importance Based on Correlation with Y:")
print(correlation)

#Plot correlation values
plt.figure(figsize=(10, 5))
sns.barplot(x=correlation.index, y=correlation.values)
plt.xticks(rotation=45)
plt.xlabel("Features")
plt.ylabel("Correlation with Y")
plt.title("Feature Importance Based on Correlation with Target (Y)")
plt.show()
```

VARIABLES THAT AFFECT TARGET LABEL

*   BMI and S5 Are the Most Important Predictors of Diabetes Progression.
*   Blood Pressure (BP) Has a Moderate Effect, Suggesting Hypertension Plays a Role.

*   Age and Sex Have Lower Correlations, Meaning Diabetes Progression Is More Metabolic-Driven.
*   Future Modeling Should Focus on High-Impact Features Like BMI and Blood Markers for Better Predictions.
